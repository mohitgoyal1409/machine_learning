```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('weight-height.csv')
```

```python
df.head()
```

```python
df['Height'].describe()
```

```python
import seaborn as sns
sns.distplot(df['Height'])
```

```python
upper_limit = df['Height'].quantile(0.99)
upper_limit

lower_limit = df['Height'].quantile(0.01)
lower_limit
```

```python
new_df = df[(df['Height'] <= 74.78) & (df['Height'] >= 58.13)]
```
