# Opening a csv file from an URL

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac 0S X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers = headers)
data = StringIO(req.text)
pd.read_csv(data)
```

# convertors

```python
def rename(name):
  if name == "No relevent experience":
       return "nre"
  else:
    return name
rename("No relevent experience")
```

```python
pd.read_csv('aug_train.csv', converters = {'relevent_experience': rename})
```
